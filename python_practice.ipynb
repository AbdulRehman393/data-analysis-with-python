# Python Annotation

Think of it as a variable and you were to put a label on a container and anything in that container should hold what container is saying.

```python
age : 32
print(age)

car : str = "Toyota"
print(car)

height : float = 6.2
print(height)

name : int = "Ali"          # Here, it's telling that int is expecting , not string
print(name)


# another way to annotate is to type comments where the interepreter will ignore it 

age = 52   # type: int
print(age)


# Any time a library is called, or an IDE works to scan your code, more computational overhead is required.
#
# Pro tip: Be strategic when annotating variables by type. This can add unnecessary overhead when overused.
#
# Type annotation is less common with Python users in data science, as it can be burdensome to manually map data
# every time a new set of data comes in. On the other hand, when doing object-oriented programming or writing functions, using type
# annotations becomes extremely important because it helps clarify code since you are
# dealing with more than just the built-in types.



#Pro tip: If a function expects a list of integers, you should annotate it as List[int],
# not just List. Being specific with your types can catch more potential bugs and misunderstandings.


import typing

List_of_numbers : typing.List[int] = [1,2,3,4]
print(List_of_numbers)
```

# typecasting

```python
# Explicit typecasting = When we manually convert a value or variable into another data type.

weight = 45.6
weight = int(weight)
print(type(weight))
print(weight)

remaining_amount = 100
remaining_amount = float(remaining_amount)
print(remaining_amount)

name = "Ahmad"
name = bool(name)
print(name)

value = True
print(type(value))
value = str(value)
print(type(value))

name = "Hammad"
name = list(name)
print(name)


# Implicit typecasting = When a value or varible converting into different data type automatically by the interpreter.

x = 5
y = 2.5

x = x/y
print(x)


# isinstance(x, int) = This checks if x is an integer or subtype of integer
# You can even check against multiple types

price = "house"

if isinstance(price, (int, float)):
    print("Valid numberi value for price")
else:
    print("Invalid value for price. Expected a number.")


```

# Control Flow

```python
# if_elif_else


marks = float(input("Enter your marks:"))

if marks >= 80:
    print("You got an A grade")
elif marks >= 70:
    print("You got a B grade")
elif marks >= 60:
    print("You got a C grade")
elif marks >= 50:
    print("You got a D grade")
else:
    print("You got an F grade")
```

```python
# Boolean in Conditional

online = True

if online:
    print("The user is online.")
else:
    print("The user is offline.")
```

```python
# Logical Operators = used on conditional operators

is_raining = True
umbrella = False

if is_raining and umbrella:
    print("You can go outside.")


if is_raining or umbrella:
    print("You can go outside.")

if not is_raining:
    print("You can go outside.")
```

```python
# Conditional Expression 

# X if condition else Y

age = 18

result = "Adult" if age >= 18 else "Minor"
print(result)


marks = 80

print("Pass" if marks >=50 else "Fail")
```

# String Methods

```python
name = input("Enter your name:")

# First occurence of a given character in a string, the postion
result = name.find("a")
print(result)

# Last occurence of a given character in a string, the position
result = name.rfind("a")

# This method turns only the very first character of the
# entire string into uppercase. Everything else in the string is forced into lowercase.
result = name.capitalize()
print(result)

# This method turns the first letter of every
# single word into uppercase, and makes the rest of the letters in those words lowercase.
result = name.title()
print(result)


# turns all the characters in a string into uppercase
result = name.upper()
print(result)

# turns all the characters in a string into lowercase
result = name.lower()
print(result)

# isdigit method will return True if all characters in the string are digits, otherwise it will return False
result = name.isdigit()
print(result)

# isalpha method will return True if all character in the string are alphabetic, otherwise it will return False
result = name.isalpha()
print(result)

# isalnum method will return True if all characters in the string are alphanumeric, otherwise it will return False
result = name.isalnum()
print(result)


# let's check how many specific characters are in the string
value = input("Enter your contact number:")
result = value.count("-")
print(result)

# replace method will replace all occurence of a specific characte with another character
result = value.replace("-","")
print(result)




```
