# Python Annotation

Think of it as a variable and you were to put a label on a container and anything in that container should hold what container is saying.

```python
age : 32
print(age)

car : str = "Toyota"
print(car)

height : float = 6.2
print(height)

name : int = "Ali"          # Here, it's telling that int is expecting , not string
print(name)


# another way to annotate is to type comments where the interepreter will ignore it 

age = 52   # type: int
print(age)


# Any time a library is called, or an IDE works to scan your code, more computational overhead is required.
#
# Pro tip: Be strategic when annotating variables by type. This can add unnecessary overhead when overused.
#
# Type annotation is less common with Python users in data science, as it can be burdensome to manually map data
# every time a new set of data comes in. On the other hand, when doing object-oriented programming or writing functions, using type
# annotations becomes extremely important because it helps clarify code since you are
# dealing with more than just the built-in types.



#Pro tip: If a function expects a list of integers, you should annotate it as List[int],
# not just List. Being specific with your types can catch more potential bugs and misunderstandings.


import typing

List_of_numbers : typing.List[int] = [1,2,3,4]
print(List_of_numbers)
```

# String Methods

```python
name = input("Enter your name:")


# to find the lenght of the string
result = len(name)
print(result)

# First occurence of a given character in a string, the postion
result = name.find("a")
print(result)

# Last occurence of a given character in a string, the position
result = name.rfind("a")

# This method turns only the very first character of the
# entire string into uppercase. Everything else in the string is forced into lowercase.
result = name.capitalize()
print(result)

# This method turns the first letter of every
# single word into uppercase, and makes the rest of the letters in those words lowercase.
result = name.title()
print(result)


# turns all the characters in a string into uppercase
result = name.upper()
print(result)

# turns all the characters in a string into lowercase
result = name.lower()
print(result)

# isdigit method will return True if all characters in the string are digits, otherwise it will return False
result = name.isdigit()
print(result)

# isalpha method will return True if all character in the string are alphabetic, otherwise it will return False
result = name.isalpha()
print(result)

# isalnum method will return True if all characters in the string are alphanumeric, otherwise it will return False
result = name.isalnum()
print(result)


# let's check how many specific characters are in the string
value = input("Enter your contact number:")
result = value.count("-")
print(result)

# replace method will replace all occurence of a specific characte with another character
result = value.replace("-","")
print(result)

# only removes spaces at the beginning and end of the string 
result = name.strip()
print(result)


# only removes spaces at the beginning of the string
result = name.lstrip()

# only removes spaces at the end of the string
result = name.rstrip()


# The .split() method will split a string into a list of substrings based on a specified delimiter. By default, it splits on whitespace.
text = "Hello World! Welcome to Python programming."
result = text.split()
print(result)


print(" ".join(result))


text = "A-quick-brown-fox-jumps-over-the-lazy-dog"
print(text.split("-"))


print("-".join(text.split( )))

# If you want to comprehensive list to all the string methods available to you
print(help(str))
```

# String Indexing 

Accessing elements of a sequence using [] (indexing operator)
[start: end : step]
step = if you want to skip over the characters


```python
credit_number = "1234-5678-9012-3456"

print(credit_number[0])
print(credit_number[:4])     # Python will asssume the starting index is 0 if we don't provide it, 
                             # so it will give us the first 4 characters of the string
print(credit_number[4:9])    # the ending string is exclusive
print(credit_number[5:])     # Everyting besides 4 digits to the end of the string
print(credit_number[-4:])    # last 4 characters of the string
print(credit_number[-3])     # if you need character in a string from the end, you can use negative indexing, 
                             # -1 will give you the last character of the string, -2 will give you the second last character and so on.
print(credit_number[::2])    # this will print every second charecter of the string

print(credit_number[2:-1:3])  # this will print every third character of the string


# if you want to reverse the  string
print(credit_number [::-1])

```

# Format Specifiers

{value:flags} when used in the context of f-strings, they allow us to format a value based on what flags are inserted


```python
price1 = 1233.372113721
price2 = -73127.65
price3 = 1200.34


# .(number)f  = round to that many decimal places (fixed point)
print(f"Price1 if {price1:.2f}")


# :(number) = Reserve a minimum width of that many characters.
# Python will never cut off the value if the value itself is longer.
print(f"Price 2 is {price2:2}")

# :03 = Allocate and zero pad that many spaces
# if you want to precede number with 0 , now these numbers will be 0 padded
print(f"Price 3 is {price3:010}")

# <: = Left Justify
print(f"Price 3 is {price3:<10}")

# >: = Right Justify
print(f"Price 3 is {price3:>10}")

# Center Align
print(f"Price 3 is {price3:^10}")


# :+ = Used to indicate a sign (+ of -)
print(f"Price1 is {price1:+}")        # Positive numbers get a + sign
print(f"Price2 is {price2:+}")        # Negative numbers keep their - sign
print(f"Price3 is {price3:+}") 


# : = or insert space
# Or Leave a blank space for positive numbers, but keep the - sign for negative numbers
print(f"Price1 is {price1: }")
print(f"Price2 is {price2: }")
print(f"Price3 is {price3: }")


# :, = comma separator 
# ther is (1,000)thousand separator which is a comma
print(f"Price1 is {price1:,}")
print(f"Price2 is {price2:,}")
print(f"Price3 is {price3:,}")

# We can also mix and match flags
print(f"Price1: {price1:,.2f}")

```
