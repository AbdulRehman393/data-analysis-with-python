## Conditional operators
`==` equal to, `!=` not equal to, `>` greater than, `less than` less than,
`>=` greater than or equal to, `<=` less than or equal to  


```python
number_of_boys = 10
number_of_girls = 5
```

```python
print(number_of_boys == number_of_girls)

print(number_of_boys == number_of_girls)

print(number_of_boys > number_of_girls)

print(number_of_boys < number_of_girls)

print(number_of_boys >= number_of_girls)

print(number_of_boys <= number_of_girls)

print(f"Are boys more in numbers? {number_of_boys > number_of_girls}")
```

## Logical Operators

- and = True when both conditions are true, else false
- or = True if any one of the condidtion is true, else false
- not = returns True if condition is False and True if condition is False


```python
today_weather = 'sunny'
tomorrow_weather = 'warm'

print(today_weather == 'sunny' and tomorrow_weather == 'sunny')

print(today_weather == 'sunny' or tomorrow_weather == 'sunny')

print(not today_weather == 'sunny')


```

The `.format` method combines values of other data types, e.g., integers, floats, booleansm lists, etc. with strings. You can use `format` to construct output message for display.

```python
# Input variables
cost_of_ice_bag = 1.25
profit_margin = .2
number_of_bags = 500

# Template for output message
output_template = """If a grocery store sells ice bags at $ {} per bg, with a profit margin of
{} %, then the total profit it makes by selling {} ice bags is $ {}."""

print(output_template)
```

```python
# Inserting values into the string
total_profit = cost_of_ice_bag * profit_margin * number_of_bags
output_message = output_template.format(cost_of_ice_bag, profit_margin*100, number_of_bags, total_profit)

print(output_message)
```

Notice how the placeholders `{}` in the `output_template` string are replaces with the arguments proveided to the `.format` method
